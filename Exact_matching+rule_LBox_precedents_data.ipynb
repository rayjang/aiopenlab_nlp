```python
from kss import split_sentences
with open('./data/precedent-crawling/case.csv','w') as f:
    for i in range(len(data_corpus['train'])):
        text = data_corpus['train'][i]['precedent'].replace('\n',' ')
        sentences = split_sentences(text)
        for s in sentences:
            f.write(s)
            f.write('\n')
```

```python
from kss import split_sentences
with open('./data/precedent-crawling/case.txt','r', encoding='utf-8') as f:
    i = 0
    while True:
        line = f.readline().replace('\n', ' ').replace('[', ' ').replace(']', ' ').replace('{', ' ').replace('}', ' ')
        sentences = split_sentences(text=line, use_heuristic=False, backend='mecab', num_workers=-1)
        if i == 0:
            output = open('./data/precedent-crawling/case-preprocessed2.txt','w', encoding='utf-8')
            for s in sentences:
                output.write(s)
                output.write('\n')
            output.close()
        else:
            output = open('./data/precedent-crawling/case-preprocessed2.txt','a', encoding='utf-8')
            for s in sentences:
                output.write(s)
                output.write('\n')
            output.close()
        i += 1 
```

```python
#postprocessing: delete Josa from extracted words(person, organization) by using kiwipiepy

output = open('./preprocessed_ner_data4.txt', 'w')
output.write('')
output.close()

from kiwipiepy import Kiwi
kiwi = Kiwi()
exclusion_tags = ['JKS','JKC','JKG','JKO','JKB','JKV','JKQ','JX','JC']
# for oneexample in test:
fx = lambda x: x in exclusion_tags

with open('./data/precedent-crawling/case-preprocessed2.txt', 'r') as f:
    while True:
        line = f.readline().replace('\n', '').replace('\\', '').replace('-', ' ').replace('[', ' ').replace(']', ' ').replace('{', ' ').replace('}', ' ').replace('(', ' ').replace(')', ' ')
        # print(line)
        if not line:
            break
        line = line.split(" ")
        # print(line)
        new_line = []
        for word in line:
            tag=0
            word_ = [x.form for x in kiwi.tokenize(word) if fx(x.tag) is False]
            word_ = " ".join(word_)
            for n in name:
                if n in word_:
                    new_line.append("["+word+"](PER)")
                    tag += 1
                    
            for o in org:
                if o in word and tag ==0:
                    print(word)
                    print(o)
                    print("------")
                    new_line.append("["+word+"](ORG)")
            if tag == 0:    
                new_line.append(word)
        # print(new_line)
        line = " ".join(new_line)
        # print(line)
        output = open('./preprocessed_ner_data4.txt', 'a')
        output.write(line)
        output.write('\n')
        output.close()
# [암유전자치료](keyword_type) 기술 관련 [다년도 과제](obj_type)?    
```
